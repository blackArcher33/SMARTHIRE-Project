```python
# Count number of skills required for each job
skill_count_df = (
    job_skills_df
    .groupby('job_id')          # Group data by job ID
    .size()                     # Count number of skills per job
    .reset_index(name='skill_count')  # Convert result to DataFrame with column name
)



skill_count_df
```

# **creating industry_count**

```python
# Count number of job postings for each industry
industry_count = df4['industry_id'].value_counts()

# Map industry-wise job count to each row as industry_popularity
df4['industry_popularity'] = df4['industry_id'].map(industry_count)


industry_count
```

# **creating job_age_day**s

```python

# Convert original_listed_time column to numeric (invalid values become NaN)
df4["original_listed_time"] = pd.to_numeric(
    df4["original_listed_time"],
    errors="coerce"
)

# Convert timestamp (in milliseconds) to datetime format
df4["original_listed_time"] = pd.to_datetime(
    df4["original_listed_time"],
    unit="ms",
    errors="coerce"
)

# Set reference date as the latest job posting date in the dataset
reference_date = df4["original_listed_time"].max().normalize()

# Create a new feature: job age in days
df4["job_age_days"] = (
    reference_date - df4["original_listed_time"].dt.normalize()
).dt.days

# Replace missing values with 0 and ensure no negative values
df4["job_age_days"] = df4["job_age_days"].fillna(0).clip(lower=0)

# Display sample values
print(df4[["original_listed_time", "job_age_days"]].head(10))

# Print number of unique job age values
print("Unique job_age_days:", df4["job_age_days"].nunique())
```

# **Data Preprocessing ( Fill Missing Value)**

“Missing values in the remote_allowed feature were treated as ‘No’, since remote work is typically explicitly stated when available.”

```python
# Convert remote_allowed values to binary format (Yes/True -> 1, No/False -> 0)
df_reduced['remote_allowed'] = df_reduced['remote_allowed'].map({
    'Yes': 1,
    'No': 0,
    True: 1,
    False: 0
})

# Fill missing remote_allowed values with 0 (assume remote work not allowed)
df_reduced['remote_allowed'] = df_reduced['remote_allowed'].fillna(0)




```

# **Filling null values with 0**

```python
# Fill missing company_size values with 0
df_reduced['company_size'] = df_reduced['company_size'].fillna(0)

# Fill missing industry_popularity values with 0
df_reduced['industry_popularity'] = df_reduced['industry_popularity'].fillna(0)

# Fill missing skill_count values with 0
df_reduced['skill_count'] = df_reduced['skill_count'].fillna(0)

# Fill missing views values with 0
df_reduced['views'] = df_reduced['views'].fillna(0)

# Fill missing applies values with 0
df_reduced['applies'] = df4['applies'].fillna(0)
```

# **Filling missing values with 'Unknown'**

```python
# Fill missing values in experience level with 'Unknown'
df_reduced['formatted_experience_level'] = (
    df_reduced['formatted_experience_level']
    .fillna('Unknown')
)

# Fill missing values in compensation type with 'Unknown'
df_reduced['compensation_type'] = df_reduced['compensation_type'].fillna('Unknown')
```

```python
# Fill missing industry_id values with -1 (represents unknown industry)
df_reduced['industry_id'] = df_reduced['industry_id'].fillna(-1)

# Convert industry_id column to integer type
df_reduced['industry_id'] = df_reduced['industry_id'].astype(int)

```

```python
# Create binary flags to indicate missing values in employee_count and follower_count
df_reduced['employee_count_missing'] = df_reduced['employee_count'].isnull().astype(int)
df_reduced['follower_count_missing'] = df_reduced['follower_count'].isnull().astype(int)

# Fill missing employee_count values using median
df_reduced['employee_count'] = df_reduced['employee_count'].fillna(
    df_reduced['employee_count'].median()
)

# Fill missing follower_count values using median
df_reduced['follower_count'] = df_reduced['follower_count'].fillna(
    df_reduced['follower_count'].median()
)
```

```python
# Create a new binary column to mark missing salary values (1 = missing, 0 = not missing)
df_reduced['salary_missing'] = df_reduced['normalized_salary'].isna().astype(int)

# Fill missing salary values using median salary
# Grouped by industry and experience level (more specific and accurate)
df_reduced['normalized_salary'] = df_reduced.groupby(
    ['industry_id', 'formatted_experience_level']
)['normalized_salary'].transform(
    lambda x: x.fillna(x.median())
)

# If some values are still missing, fill them using median salary
# Grouped only by experience level (broader fallback)
df_reduced['normalized_salary'] = df_reduced.groupby(
    'formatted_experience_level'
)['normalized_salary'].transform(
    lambda x: x.fillna(x.median())
)
```

# **Random Forest Parameter Tunning**

```python
# Import Random Forest Regressor and RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV

# Create base Random Forest Regressor model
rf = RandomForestRegressor(
    random_state=42,   # Fix random seed for reproducibility
    n_jobs=-1          # Use all available CPU cores
)

# Define hyperparameter search space for Random Forest
param_dist = {
    'n_estimators': [200, 300, 400, 500],   # Number of trees in the forest
    'max_depth': [None, 10, 20, 30, 40],    # Maximum depth of each tree
    'min_samples_split': [2, 5, 10],        # Minimum samples required to split a node
    'min_samples_leaf': [1, 2, 4],          # Minimum samples required at a leaf node
    'max_features': ['sqrt', 'log2']        # Number of features considered at each split
}

# Create RandomizedSearchCV object for hyperparameter tuning
rf_random = RandomizedSearchCV(
    estimator=rf,                # Base Random Forest model
    param_distributions=param_dist, # Hyperparameter combinations
    n_iter=20,                    # Try 20 random combinations
    cv=5,                         # 5-fold cross-validation
    scoring='r2',                 # Use R2 score for evaluation
    random_state=42,              # Fix random seed
    n_jobs=-1,                    # Use all CPU cores
    verbose=2                     # Show training progress
)

# Fit RandomizedSearchCV on full dataset
rf_random.fit(X, y)

# Print best hyperparameters found
print("Best Parameters:", rf_random.best_params_)

# Print best cross-validation R2 score
print("Best CV R2:", rf_random.best_score_)

```

```python
# Get the best Random Forest model after hyperparameter tuning
best_rf = rf_random.best_estimator_

# Predict target values on the test dataset using tuned Random Forest model
y_pred_tuned = best_rf.predict(X_test)

# Print Mean Absolute Error for tuned Random Forest model
print("Tuned RF MAE:", mean_absolute_error(y_test, y_pred_tuned))

# Print R2 score for tuned Random Forest model
print("Tuned RF R2:", r2_score(y_test, y_pred_tuned))
```

# **XGBRegressor using Hyperparameter Tuninng**

```python
from xgboost import XGBRegressor
from sklearn.model_selection import RandomizedSearchCV
```

```python
# Define hyperparameter search space for XGBoost model
param_dist_xgb = {
    'n_estimators': [200, 300, 400, 500],   # Number of trees in the model
    'learning_rate': [0.01, 0.05, 0.1],     # Step size used to update weights
    'max_depth': [3, 5, 7, 9],              # Maximum depth of each tree
    'subsample': [0.6, 0.8, 1.0],            # Fraction of training data used for each tree
    'colsample_bytree': [0.6, 0.8, 1.0],     # Fraction of features used for each tree
    'min_child_weight': [1, 3, 5]            # Minimum sum of instance weight needed in a child
}

# Create XGBoost Regressor model
xgb = XGBRegressor(
    objective='reg:squarederror',  # Regression objective (used for continuous target)
    random_state=42                # Fix random seed for reproducibility
)
```

# hyperparameter tunning of  XGBRegressor using RandomizedSearchCV

```python
# Create RandomizedSearchCV object for XGBoost hyperparameter tuning
xgb_random = RandomizedSearchCV(
    estimator=xgb,                    # Base XGBoost model
    param_distributions=param_dist_xgb, # Hyperparameter search space
    n_iter=20,                         # Number of random parameter combinations to try
    cv=5,                              # 5-fold cross-validation
    scoring='r2',                      # Use R2 score to evaluate model performance
    random_state=42,                   # Fix random seed for reproducibility
    n_jobs=-1,                         # Use all available CPU cores
    verbose=2                          # Show progress during training
)

# Fit RandomizedSearchCV on full dataset to find best parameters
xgb_random.fit(X, y)
```
